SIFT - Scale invarient feature transform

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('E:/python/Images for opencv/Cat image.jpg')
gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
plt.imshow(gray)

sift = cv2.SIFT_create()

keypoints = sift.detect(gray , None)
print("number of keypoints detected : " , len(keypoints))

image = cv2.drawKeypoints(image , keypoints ,None , flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(image)
```

SURF - Speeded up robust features

```python
image2 = cv2.imread('E:/python/Images for opencv/Cat image.jpg')
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
plt.imshow(gray2)

surf = cv2.xfeatures2d.SURF_create()

surf.hessianThreshold = 7500
keypoints2 , descriptors = surf.detectandcompute(gray2 , None)
print("Number of keypoints detected : " , len(keypoints2))

image2 = cv2.drawKeypoints(image2 , keypoints2 , None , flags = cv2.DRAW_MATCHES_FLAGs_DRAW_RIC_POINTS)
plt.imshow(image2)
```
