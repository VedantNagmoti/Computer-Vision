Arithematic operations

this operations allow us to directly add or subtract to the colour intensity

calculates the per elment operation of two arrays

the overall effect is incresing or decresing the brightness

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('E:/python/Images for opencv/Cat image.jpg')
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
```

Create a matrix of ones and multiply it with scaler of 100

it gives matrix with same dimensions as of our image with all values being 100

```python
M = np.ones(image.shape , dtype="uint8") * 100
print(M)
```

```python
# Add matrix to image
added = cv2.add(image , M)
plt.imshow(added)
```

```python
subtracted = cv2.subtract(image , M)
plt.imshow(subtracted)
```
