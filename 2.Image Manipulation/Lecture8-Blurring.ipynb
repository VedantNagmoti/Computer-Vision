BLURRING

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('E:/python/Images for opencv/Cat image.jpg')
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

1. cv2.filter2D(image , -1 , kernel)

```python
# Creating our 3x3 kernel
kernel1 = np.ones((3 , 3) , np.float32) / 9         # divided by 9 because 9 = 3 x 3
blurred1 = cv2.filter2D(image , -1 , kernel1)
plt.imshow(blurred1)
```

```python
# Creating our 7x7 kernel
kernel2 = np.ones((7 , 7) , np.float32) / 49
blurred2 = cv2.filter2D(image , -1 , kernel2)
plt.imshow(blurred2)
```

```python
# Creating our 20x20 kernel
kernel3 = np.ones((20 , 20) , np.float32) / 400
blurred3 = cv2.filter2D(image , -1 , kernel3)
plt.imshow(blurred3)
```

Other commonly used blurring methodes used in openCV


cv2.blur(image , (3 , 3))

```python
blur = cv2.blur(image , (20 , 20))
plt.imshow(blur)
```

```python
gaussian = cv2.GaussianBlur(image , (7 , 7) , 0)
plt.imshow(gaussian)
```

```python
median = cv2.medianBlur(image , 5)
plt.imshow(median)
```

```python
bilateral = cv2.bilateralFilter(image , 9 , 75 , 75)
plt.imshow(bilateral)
```
