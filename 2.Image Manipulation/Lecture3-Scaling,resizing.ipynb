```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Scaling  , Resizing , interpolations

cv2.resize(image , output image size , x scale , y scale , interpolation)

```python
image = cv2.imread('E:/python/Images for opencv/Cat image.jpg')
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
quarter_image = cv2.resize(image , None , fx=0.25 , fy=0.25)            #without mentioning interpolation
plt.imshow(quarter_image)
```

```python
double_image = cv2.resize(image , None , fx = 2 , fy = 2 , interpolation=cv2.INTER_CUBIC )
plt.imshow(double_image)
```

```python
image_scaled = cv2.resize(image , (600 , 600) , interpolation=cv2.INTER_AREA)
plt.imshow(image_scaled)
```
