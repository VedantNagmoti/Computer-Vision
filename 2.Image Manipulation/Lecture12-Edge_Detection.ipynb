Edge detection and image gradients

Edge detection algorithms

Sobel - to identify vertical or horizontal edges

laplacian - gets all orientations

Canny - optimal due to low error rate , well defined edges and Accurate detection


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Canny edge detection algorithm

1. Applies gaussian blurring
2. finds intensity gradient of the image
3. Applid non -  maximum suppression (i.e removes pixels that are not edges)
4. Hysterisis - Applies threshholds , pixels above or below certain threshholds it considers as edge


```python
image = cv2.imread('E:/python/Images for opencv/Cat image.jpg')
image = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
plt.imshow(image , cmap="grey")
```

```python
height , width = image.shape
print("height = ",height , "\nwidth = " , width)
```

```python
sobel_x = cv2.Sobel(image , cv2.CV_64F , 0 , 1 , ksize=5)
sobel_y = cv2.Sobel(image , cv2.CV_64F , 1 , 0 , ksize=5)
```

```python
plt.imshow(sobel_y , cmap="grey")
```

```python
plt.imshow(sobel_x , cmap="grey")
```

```python
sobel_or = cv2.bitwise_or(sobel_x , sobel_y)
plt.imshow(sobel_or , cmap="grey")
```

```python
laplacian = cv2.Laplacian(image , cv2.CV_64F)
plt.imshow(laplacian , cmap="grey")
```

```python
canny = cv2.Canny(image , 20 , 200)     #you can change thresholds as per image seems nice

plt.imshow(canny , cmap="grey")
```
