Cropping


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('E:/python/Images for opencv/Cat image.jpg')
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
height , width = image.shape[:2]
start_x , start_y = int(height * 0.25) ,int( width * 0.25) # Starting co-ordinate
end_x , end_y = int(height * 0.75), int(width * 0.75) # End co-ordinate
cropped = image[start_x:end_x, start_y:end_y]
```

```python
plt.imshow(cropped)
```
