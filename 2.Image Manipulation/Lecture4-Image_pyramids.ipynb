Image Pyramids

Usefull when scaling images in object detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('E:/python/Images for opencv/Cat image.jpg')
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
smaller = cv2.pyrDown(image)        #half the size
bigger = cv2.pyrUp(image)           #Twice the size
# plt.imshow(image)
plt.imshow(smaller)
# plt.imshow(bigger)
```
