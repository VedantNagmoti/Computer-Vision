THRESHOLDING

Everything above the certain threshold value will be converted to black and everything below to white.

used for binerization

Types of Thresholding

cv2.THRESH_BINERY   - Most common

cv2.THRESH_BINERY_INV  - Most common

cv2.THRESH_TRUNC

cv2.THRESH_TOZERO

cv2.THRESH_TOZERO_INV

NOTE :- Images need to be converted to the greyscale image before thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

cv2.threshhold(image , threshold value , max pixel value , threshhold type)

```python
image = cv2.imread('E:/python/Images for opencv/Cat image.jpg')
image = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
plt.imshow(image , cmap = "grey")
```

```python
# below 127 = 0
# above 127 = 255
ret,thresh1 = cv2.threshold(image , 127 , 255 , cv2.THRESH_BINARY) 
plt.imshow(thresh1 , cmap="grey")
```

```python
# below 127 = 255
# above 127 = 0
ret,thresh1 = cv2.threshold(image , 127 , 255 , cv2.THRESH_BINARY_INV) 
plt.imshow(thresh1 , cmap="grey")
```

```python
# below 127 = 0
# above 127 = unchanged
ret,thresh1 = cv2.threshold(image , 127 , 255 , cv2.THRESH_TOZERO) 
plt.imshow(thresh1 , cmap="grey")
```

```python
# below 127 = unchanged
# above 127 = 255
ret,thresh1 = cv2.threshold(image , 127 , 255 , cv2.THRESH_TOZERO_INV) 
plt.imshow(thresh1 , cmap="grey")
```

```python
# below 127 = 0
# above 127 = truncated (held) at 127
ret,thresh1 = cv2.threshold(image , 127 , 255 , cv2.THRESH_TRUNC) 
plt.imshow(thresh1 , cmap="grey")
```

Adaptive thresholding

cv2.adaptiveThreshold(image , max value , adaptive type , threshold type , block size , constant that is subtracted from mean)

Note : block size need to be odd number

Adaptive threshold types

ADAPTIVE_THRESH_MEAN_C - BASED ON THE MEAN OF THE NEIGHBOURHOOD PIXELS

ADAPTIVE_THRESH_GAUSSIAN_C - weigted sum of neighborhood pixels under the gaussian window

THRESH_OTSU

```python
thresh2 = cv2.adaptiveThreshold(img , 255 , cv2.ADAPTIVE_THRESH_MEAN_C , cv2.THRESH_BINARY , 3 ,5 )
plt.imshow(thresh2 , cmap="grey")
```

```python
_, th2 = cv2.threshold(img , 0 , 255 , cv2.THRESH_BINARY + cv2.THRESH_OTSU )
plt.imshow(th2 , cmap="grey")
```
