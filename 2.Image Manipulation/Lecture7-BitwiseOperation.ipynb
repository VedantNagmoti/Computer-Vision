BITWISE OPERATIONS

Creating a greyscale image 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# making square
square = np.zeros((300 , 300) , np.uint8)
cv2.rectangle(square , (50 , 50) , (250 , 250) , 255 , -2)
plt.imshow(square , cmap="grey")
```

```python
# Elipse
elipse = np.zeros((300 , 300) , np.uint8)
cv2.ellipse(elipse , (150 , 150) , (150 , 150) , 30 , 0 , 180 , 255 , -1)
plt.imshow(elipse , cmap="grey")
```

```python
cv2.imshow("Square",square)
cv2.imshow("Ellipse",elipse)
cv2.waitKey()
cv2.destroyAllWindows()
```

EXPERIMENTING WITH SOME BITWISE OPERATIONS

```python
AND = cv2.bitwise_and(square , elipse)
plt.imshow(AND)
```

```python
OR = cv2.bitwise_or(square , elipse)
plt.imshow(OR)
```

```python
XOR = cv2.bitwise_xor(square , elipse)
plt.imshow(XOR)
```

```python
NOT = cv2.bitwise_not(square)
plt.imshow(NOT)
```
