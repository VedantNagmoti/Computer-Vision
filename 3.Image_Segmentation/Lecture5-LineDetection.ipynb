cv2.Houghlines(binerized image , p accuracy , theta acuracy , threshold) Minimum line length maximmum line gap

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('E:/python/Images for opencv/sudoku.jpg')
plt.imshow(image)
```

```python
grey = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
plt.imshow(grey , cmap = "grey")
```

```python
edges = cv2.Canny(grey , 100 , 175 , apertureSize=3)
plt.imshow(edges)
```

```python
# run hough lines as an rho accuracy = 1 pixel
# theta accuracy = np.pi / 180    i.e. 1 degree
#line threshhold = 140 (Number of points on the line)
lines = cv2.HoughLines(edges , 1 , np.pi/180 , 140)

# Iterate through each line and convert it to the format required by cv2.lines i.e to get required end points
if lines is not None:
    for line in lines:
        rho, theta = line[0]  # Extract rho and theta

        # Convert polar coordinates to Cartesian
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * a)
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * a)

        # Draw the detected line
        cv2.line(image, (x1, y1), (x2, y2), (255, 0, 0), 2)

# Show Image
plt.imshow(image)
```

```python
image2 = cv2.imread('E:/python/Images for opencv/sudoku.jpg')
grey2 = cv2.cvtColor(image2 , cv2.COLOR_BGR2GRAY)
edges2 = cv2.Canny(grey2 , 100 , 175 , apertureSize=3)
```

```python
lines2 = cv2.HoughLinesP(edges2 , 1 , np.pi/180 , 200 , 5 , 10)
print(lines2.shape)

for x1 , x2 , y1 , y2 in lines2[0] :
    cv2.line(image2 , (x1, y1) , (x2 , y2) , (0 , 255 ,0) , 3)

plt.imshow(image2)
```
