```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('E:/python/Images for opencv/shapes.png')
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
plt.imshow(gray , cmap="gray")
```

```python
edged = cv2.Canny(gray , 50 , 200)
plt.imshow(edged , cmap="gray")
```

```python
blank_image = np.zeros((image.shape[0], image.shape[1] , 3))
plt.imshow(blank_image)
```

```python
contours , hierarchy = cv2.findContours(edged.copy() , cv2.RETR_EXTERNAL , cv2.CHAIN_APPROX_NONE)
print("number of contours found : ", len(contours))
```

```python
cv2.drawContours(blank_image , contours , -1 , (0 , 255 , 0) , 3)
plt.imshow(blank_image)
```

```python
cv2.drawContours(image , contours , -1 , (0 , 255 , 0) , 3)
plt.imshow(image)
```

Sorting contour by area


```python
def get_contour_area(contours) :
    all_areas = []
    for cnt in contours:
        area = cv2.contourArea(cnt)
        all_areas.append(area)
    return all_areas
```

```python
image = cv2.imread('E:/python/Images for opencv/shapes.png')
original = image
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
print("Contours area before sorting : " ,get_contour_area(contours))
```

```python
sorted_contours = sorted(contours , key=cv2.contourArea , reverse=True)
```

```python
print("Contour area after sorting : " , get_contour_area(sorted_contours))
```

```python
for i in sorted_contours:
    cv2.drawContours(image , [i] , -1 , (0 , 255 , 0) , 3)
    cv2.waitKey()
    cv2.imshow("contours by area" , image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
