BLOB DETECTION

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('E:/python/Images for opencv/sunflower.png')
gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
params = cv2.SimpleBlobDetector_Params()

# Optional: Customize parameters (e.g., filter by color, size, etc.)
params.filterByArea = True
params.minArea = 100  # Minimum area of blobs
params.filterByCircularity = False  # Set to True if you want circular blobs
params.filterByConvexity = False
params.filterByInertia = False

# Create a blob detector
detector = cv2.SimpleBlobDetector_create(params)

# Detect blobs
keypoints = detector.detect(gray)

# Draw blobs
blank = np.zeros((1, 1))
blobs = cv2.drawKeypoints(image, keypoints, blank, (0, 255, 0), cv2.DRAW_MATCHES_FLAGS_DEFAULT)

# Convert image to RGB for Matplotlib display
blobs_rgb = cv2.cvtColor(blobs, cv2.COLOR_BGR2RGB)

plt.imshow(blobs_rgb)
```
