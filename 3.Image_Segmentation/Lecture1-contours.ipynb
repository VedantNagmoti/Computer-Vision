```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('E:/python/Images for opencv/contours.png')
cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
plt.imshow(image , cmap="gray")
```

```python
#find canny edges
edged = cv2.Canny(image , 30 , 200)
plt.imshow(edged , cmap="gray")
```

```python
# finding contours
contours , hierarchy = cv2.findContours(edged , cv2.RETR_EXTERNAL , cv2.CHAIN_APPROX_NONE)
plt.imshow(edged , cmap = "gray")
```

```python
print("number of contours found : " , str(len(contours)))
```

```python
cv2.drawContours(image , contours , -1 , (0 , 255 , 0) , 3)
plt.imshow(image)
```

cv2.findContours(image , retrievel mode , approximation method)

Approximation methods
1. v2.CHAIN_APPROX_NONE => stores all the boundary points , but we dont necessarily need all the bounding points , if the points form a straight line , we only need the start and the end points of the straight line

2. cv2.CHAIN_APPROX_SIMPLE  =>  Only provides the start and the end points of bounding contours , thus efficient storage of contour information

Retrievel Nodes - (Hierarchy in contours)

heirarchy types (the first two are the most usefull)
1. cv2.RETR_LIST  - retrieves all contours
2. cv2.RETR_EXTERNAL - retrieves external or outer countours only
3. cv2.RETR_CCOMP - Retrieves all in 2 level hierarchy
4. cv2.RETR_TREE - Retrieves all in full hierarchy

hierarchy is stored in the following format - [next , previous , fist child , parent]
