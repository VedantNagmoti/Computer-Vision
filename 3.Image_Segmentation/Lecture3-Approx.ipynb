cv2.approxPolyDP(contour , Approximation accuracy , closed)

Approximation accuracy => small values give precise approximation , large values give more generic approximation , should be less than 5% of contour perimeter

Closed => boolean value to state approximation should be closed or opened

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('E:/python/Images for opencv/approx.png')
plt.imshow(image)
```

```python
original = image.copy()
grey = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)

contours , hierarchy = cv2.findContours(grey , cv2.RETR_LIST , cv2.CHAIN_APPROX_NONE)
```

```python
# Drawing bounding rectangle 
for c in contours :
     x , y , w , h = cv2.boundingRect(c)
     cv2.rectangle(image , (x , y) , (x+w, y+h) , (0 , 0 ,255) , 2)
     
plt.imshow(image)
```

```python
for c in contours :
    accuracy = 0.03 * cv2.arcLength(c , True)
    approx = cv2.approxPolyDP(c , accuracy , True)
    cv2.drawContours(original , [approx] , 0 , (0 , 255 , 0) , 2)
    
plt.imshow(original)
```
