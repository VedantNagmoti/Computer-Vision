Drawing Images and Shapes Using OpenCV

```python
import cv2
import numpy as np
```

Creating a black square


```python
image = np.zeros((512 , 512 , 3) , np.uint8)        # 3 Layered [RGB] Black Image
image2 = np.zeros((512 , 512) , np.uint8)           # 2 Layered [Black and White] Image

cv2.imshow("Coloured Black image" , image)
cv2.imshow("Black and white image" , image2)
cv2.waitKey()
cv2.destroyAllWindows()
```

Let's Draw a blue line over Black image

Syntax : 

cv2.line(image , (x1 , y1) , (x2 , y2), Colour in BGR Format (255 , 177 , 0) , Thickness in pixels 5 )

```python
cv2.line(image , (0 , 0) , (511 , 511) , (255 , 177 , 0) , 5 )
cv2.imshow("Blue line image" , image)
cv2.waitKey()
cv2.destroyAllWindows()
```

Rectangle

cv2.rectangle(image , (x1 , y1) , (x2 , y2) , colour , fill / thickness)

```python
image = np.zeros((600 , 600 , 3) , np.uint8)
cv2.rectangle(image , (200 , 200) , (400 , 400),( 255 , 177 , 0), -1)        # pixels = -1 will fill the whole rectangle 
cv2.imshow ("Filled rectangle" , image)
cv2.waitKey()
cv2.destroyAllWindows()
```

Circle

cv2.circle(image , Center , radius , color , fill)

```python
image = np.zeros((600 , 600 , 3), np.uint8)
cv2.circle(image , (300 , 300) , 100 , (0 , 0 , 255) , -1)
cv2.imshow("Filled circle" , image)
cv2.waitKey()
cv2.destroyAllWindows()
```

Polygons



```python
image = np.zeros((600 , 600 , 3), np.uint8)

# Let's define four points
pts = np.array([[10 , 50] , [400 , 50] , [90 , 200] , [50 , 500]] , np.int32)

# Let's now reshape our polylines required by polylines
pts = pts.reshape((-1 , 1 , 2))

cv2.polylines(image , [pts] , True , (0 , 255 , 255) , 5)
cv2.imshow("Polygon image" , image)
cv2.waitKey()
cv2.destroyAllWindows()
```

Text with cv2.putText

cv2.putText(image , "Hello world" , bottom left starting point , font , font size , colour , thickness)

FONT_HERSHEY_SIMPLEX, FONT_HERSHEY_PLAIN

FONT_HERSHEY_DUPLEX , FONT_HERSHEY_COMPLEX

FONT_HERSHEY_TRIPLEX , FONT_HERSHEY_COMPLEX_SMALL

FONT_HERSHEY_SCRIPT_COMPLEX

FONT_HERSHEY_SCRIPT_SIMPLEX

```python
image = np.zeros((600 , 600 , 3) , np.uint8)
cv2.putText(image , "Hello World" , (50 , 300) , cv2.FONT_HERSHEY_COMPLEX ,2 , (0 , 0 , 255) , 2)
cv2.imshow("Hello world",image)
cv2.waitKey()
cv2.destroyAllWindows()
```
