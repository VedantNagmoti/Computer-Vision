```python
blue , green , red = cv2.split(image)
cv2.imshow("Original Image", image)
cv2.imshow("Red Image" , red)
cv2.imshow("Blue Image" , blue)
cv2.imshow("Green Image" , green)
cv2.waitKey()
cv2.destroyAllWindows()
```

HISTOGRAM PLOTTING

cv2.calcHist(image , channels , mask , histsizes , ranges[,hist[,accumulate]])

 -Images - It is the type of uint8 or float32 , It should be given in square brackets that is: "[image]"

 -Channels - Also given in square brackets , Greyscale = [0] , color image = [0] / [1] / [2] for calulating histogram of B , G , R respectively

 -Mask - Mask image , to find histogram of full image it is given as 'none'
 
 -histsize = this represents our bin count , need to  be given in square brackets , for full scale we pass [256]
 
 -ranges = normaly this is [0 , 256]

```python
img = cv2.imread('C:/Users/vedan/OneDrive/Pictures/rr3/FORD MUSTANG GT PREMIUM.jpg')
histogram = cv2.calcHist([img] , [0] , None , [256] , [0 , 256])
plt.hist(img.ravel() , 256 , [0 , 256])     # Ravel flattens the image array , 2D array to 1D array
plt.show()
```

```python
# viewing saperate color channels
color = ('b' , 'g' , 'r')

# We now saparate the colours and plot each in the histogram
for i , col in enumerate(color):
    histogram2 = cv2.calcHist([img] , [i] , None , [256] , [0 , 256])
    plt.plot(histogram2 , color = col)
    plt.xlim([0 , 256])

plt.show()
```
