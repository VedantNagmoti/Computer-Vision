```python
img = cv2.imread("C:\\Users\\vedan\\OneDrive\\Pictures\\mahadev.jpg")
cv2.waitKey()

greyimg = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)

cv2.imshow("Greyscale image",greyimg)
cv2.waitKey()
cv2.destroyAllWindows()
```
