```python
import cv2
import numpy as np
```

```python
img = cv2.imread('C:/Users/vedan/OneDrive/Pictures/rr3/BMW M3 GTS.webp')
```

```python
# BGR values for the first 0,0 pixels
B1 ,G1 ,R1  = img[0 , 0]

# BGR values for the first 10,20 pixels
B2 ,G2 ,R2  = img[10 , 20]
print(img.shape)
print(B1 , G1 , R1)
print(B2 , G2 , R2)
```

```python
greyimg = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
print(greyimg.shape)
greyimg[0 , 0]
```

```python
hsvimg = cv2.cvtColor(img , cv2.COLOR_BGR2HSV)

cv2.imshow("Hsv image" , hsvimg)
cv2.imshow("Hue channel" , hsvimg[ :, :, 0 ])
cv2.imshow("Saturation channel" , hsvimg[ :, :, 1 ])
cv2.imshow("Value channel" , hsvimg[ :, :, 2 ])

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Individual channels in RGB image
B , G , R = cv2.split(img)
print (B.shape)
cv2.imshow("Original image",img)
cv2.imshow("Red" , R)
cv2.imshow("Blue" , B)
cv2.imshow("Green" , G)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# remaking original image
merged = cv2.merge([B , G , R])
cv2.imshow('Merged Image', merged)

# Amplifying the blue color
merged = cv2.merge([B+100 , G , R])
cv2.imshow("Amplified Image",merged)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# creating matrix of zeros
# dmensions of matrix = h x w

B , G , R = cv2.split(img)

zeros = np.zeros(img.shape[:2] , dtype = "uint8")
cv2.imshow("Red" , cv2.merge([zeros , zeros , R]))
cv2.imshow("Green" , cv2.merge([zeros , G , zeros ]))
cv2.imshow("Blue" , cv2.merge([B , zeros , zeros]))
cv2.waitKey()
cv2.destroyAllWindows()
```
